* **정렬(Sorting):** 데이터를 크기순(1등~꼴등)이나 가나다순으로 깔끔하게 정렬한다.

## 🎯 PART 1. 핵심 개념 확인

*판다스(Pandas)를 이용해 데이터를 요리하는 기본 문법과 용어들을 확인합니다.*

1. 표 형태의 데이터프레임에서 하나의 열(컬럼)만 떼어내어 가져올 때는 데이터프레임 이름 뒤에 **( 대괄호 )** 를 쓰고 그 안에 열 이름을 적습니다. (예: `df['이름']`)
2. 하나의 열을 가져오면 1차원 형태인 시리즈(Series)가 되지만, **두 개 이상** 의 열을 한 번에 가져오려면 대괄호를 두 번 겹쳐 쓴 `df[['열1', '열2']]` 형태로 리스트를 넣어주어야 합니다.
3. 파이썬 기본 문법에서 '같다'는 `==`입니다. 그렇다면 '크거나 같다(이상)'를 의미하는 비교 연산자 기호는 **( >= )** 입니다.
4. 판다스에서 특정 조건에 맞는 데이터만 걸러내는 방식을 **( Boolean ) Indexing** 이라고 부릅니다. 조건이 참(True)인 행만 남기는 원리입니다.
5. 판다스 필터링에서 두 가지 조건이 **모두 참이어야 할 때(AND)** 는 일반 파이썬의 `and` 대신 기호 **( & )**(앰퍼샌드)를 사용해야 해요.
6. 판다스 필터링에서 두 가지 조건 중 **하나만 참이어도 될 때(OR)** 는 일반 파이썬의 `or` 대신 기호 **( | )**(파이프, 보통 Enter 키 위)를 사용해야 돼요.
7. 다중 조건을 사용할 때는 연산자 우선순위 때문에 헷갈리지 않도록 반드시 각 조건을 **( () )** 로 묶어주어야 에러가 나지 않습니다.
8. 특정 문자가 포함된 데이터를 찾을 때, 엑셀의 '텍스트 포함' 필터와 같은 역할을 하는 판다스의 문자열 메서드는 `str.( contains )('찾을단어')` 입니다.
9. 데이터를 특정 열을 기준으로 정렬할 때 사용하는 판다스의 대표적인 메서드는 `( sort_values )` 입니다.
10. 데이터를 정렬할 때 기본값은 오름차순(작은 것부터 큰 것 순)입니다. 이를 **내림차순(큰 것부터 작은 것 순)** 으로 바꾸려면 정렬 함수 안에 `( ascending ) = False` 라는 옵션을 주어야 합니다.
11. 필터링이나 정렬을 하고 나면 행 번호(인덱스)가 뒤죽박죽 섞이게 됩니다. 이를 0부터 다시 깔끔하게 번호를 매겨주는 메서드는 `( reset_index )` 입니다.

* **예상 결과:** `<class 'pandas.core.series.Series'>` (힌트: 1차원 데이터)

````python
import pandas as pd

data = {
    "name": ["철수", "영희", "민수"],
    "age": [20, 22, 21],
    "score": [90, 95, 80]
}

df = pd.DataFrame(data)
print(df)

col = df['name']
print(type(col))
````

**문제 2. 다중 열(Columns) 선택하기**

```python
subset = df[['name', 'score']]
print(subset.shape)
```

* **예상 결과:** (3, 2) (힌트: 행은 3개, 열은 몇 개일까요?)


````python
subset = df[['name', 'score']]
print(subset.shape)
````

**문제 3. 단일 조건 필터링**

```python
smart_students = df[df['score'] >= 90]
print(len(smart_students))
```

* **예상 결과:** 2명 (누구누구 남을까요?)

````python
smart_students = df[df['score'] >= 90]
print(len(smart_students))
````

**문제 4. 다중 조건 필터링 (OR 조건)**

```python
result = df[(df['age'] == 20) | (df['score'] == 80)]
print(result['name'].tolist())
```

* **예상 결과:** ['철수', '민수'] (힌트: 20살이거나 성적이 80점인 사람)

````python
result = df[(df['age'] == 20) | (df['score'] == 80)]
print(result['name'].tolist())
````

**문제 5. 내림차순 정렬 (1등부터 줄 세우기)**

```python
sorted_df = df.sort_values(by='score', ascending=False)
print(sorted_df.iloc[0]['name'])
```

* **예상 결과:** 영희 (힌트: 성적이 가장 높은 사람은 맨 위(0번 인덱스)로 올라옵니다.)
---

````python
sorted_df = df.sort_values(by='score', ascending=False)
print(sorted_df.iloc[0]['name'])
````

## 🚀 PART 4. [심화 문제] VIP 타겟 마케팅 대상자 추출하기

*선택, 필터링, 정렬을 하나로 묶어 실무에서 가장 많이 하는 데이터 가공 작업을 수행해 봅니다.*

**[비즈니스 시나리오 및 요구사항]**
당신은 온라인 쇼핑몰의 마케터입니다. 신상품 VIP 초대장을 보낼 고객 명단을 뽑아야 합니다. `mall_data`라는 데이터프레임이 준비되어 있습니다. AI에게 아래 로직을 순서대로 설명하여 최종 결과물(`vip_list`)을 만드는 단 한 줄(또는 두 줄)의 코드를 받아내세요.

1. **필터링:** 회원 등급(`level`)이 'Gold' 이거나 'VIP'인 사람만 먼저 추려냅니다.
2. **필터링(AND):** 그중에서 이번 달 구매 금액(`monthly_spend`)이 50만 원 이상인 사람만 남깁니다.
3. **열 선택:** 분석에 필요 없는 개인정보는 빼고, '고객ID', '이름', 'monthly_spend' 열만 남깁니다.
4. **정렬:** 초대장을 구매 금액이 높은 사람부터 순서대로 보내기 위해, 'monthly_spend'를 기준으로 내림차순 정렬합니다.

* **✍️ 위 모든 조건을 한 번에 처리하기 위해 AI에게 내린 나의 구체적인 프롬프트:**
```text
판다스로 mall_data 데이터프레임에서 level이 'Gold' 또는 'VIP'인 행만 선택하고,
그중에서 monthly_spend가 500000 이상인 데이터만 남긴 뒤,
'고객ID', '이름', 'monthly_spend' 컬럼만 선택해서
monthly_spend 기준으로 내림차순 정렬하여 vip_list에 저장하는 코드를 짜줘.
```

* **💻 AI가 짜준 판다스 코드 완성본 적어보기:**
```python
vip_list = mall_data[
    ((mall_data['level'] == 'Gold') | (mall_data['level'] == 'VIP')) &
    (mall_data['monthly_spend'] >= 500000)
][['고객ID', '이름', 'monthly_spend']].sort_values(by='monthly_spend', ascending=False)
```
