## 🎯 PART 1. 개념 확인

*프로그램의 흐름을 제어하는 핵심 문법과 논리 연산의 개념을 확인합니다.*

1. 프로그램이 위에서 아래로만 실행되지 않고, 조건에 따라 갈래를 나누거나 특정 구간을 반복하도록 흐름을 통제하는 문법을 총칭하여 **( 제어문 )**(Control Flow)이라고 부릅니다.
2. 조건문의 뼈대가 되는 자료형으로, 오직 `True`(참) 또는 `False`(거짓) 두 가지 값만 가질 수 있는 자료형을 **( 불리언 )**(Boolean) 타입이라고 합니다.
3. 파이썬에서 두 값이 '같은지' 비교할 때는 `=`가 아니라 **( == )** 연산자를 사용하고, '다른지' 비교할 때는 `!=` 연산자를 사용합니다.
4. 여러 개의 조건을 결합할 때 사용하는 논리 연산자로, 양쪽 조건이 모두 참일 때만 True를 반환하는 연산자는 **( and )** 입니다. (예: `A and B`)
5. 파이썬은 중괄호 `{}` 대신 띄어쓰기 4칸을 사용하여 "이 코드는 if문(또는 for문)에 속해 있다"는 블록(Block) 구조를 인식합니다. 이를 **( 들여쓰기 )**(Indentation)이라고 부릅니다.
6. `if` 조건이 거짓일 때 다음으로 검사할 조건을 추가로 제시하려면 **( elif )** 키워드를 사용합니다. (`else if`의 줄임말)
7. 횟수가 정해져 있거나, 리스트 같은 순차적인 데이터(Sequence)의 요소를 처음부터 끝까지 하나씩 꺼내며 반복할 때는 **( for )** 문을 사용합니다.
8. `for`문과 자주 함께 쓰이며, 시작 숫자부터 끝 숫자 직전까지 연속된 정수 배열을 생성해 주는 내장 함수는 **( range() )** 입니다. (예: `range(1, 5)` -> 1, 2, 3, 4)
9. 특정 조건이 참(True)인 동안에는 횟수에 상관없이 무한히 반복을 수행하는 문법은 **( while )** 문입니다.
10. 반복문(For, While)을 실행하던 도중, 특정 조건을 만족했을 때 **반복문을 즉시 강제 종료**하고 빠져나오는 키워드는 **( break )** 입니다.
11. 반복문을 종료하지는 않지만, **현재 차례의 남은 코드를 건너뛰고** 다음 반복 차례로 곧바로 넘어가게 하는 키워드는 **( continue )** 입니다.

**문제 3. 반복문을 이용한 누적 합계 (Accumulator Pattern)**
알고리즘의 가장 기본이 되는 누적 합계 패턴입니다. `total`의 최종 값은?

```python
total = 0
prices = [100, 200, 300]
for p in prices:
    total = total + p
print(total)
```

* **예상 결과:** 600 (이유: 리스트 prices에 들어있는 세 개의 숫자(100, 200, 300)를 for문을 통해 순차적으로 total 변수에 더해나갔기 때문입니다.)

````python
total = 0
prices = [100, 200, 300]
for p in prices:
    total = total + p
print(total)
````

**문제 4. while문과 무한 루프 제어 (break)**
`while True:`는 무한 루프를 만듭니다. 코드가 언제 멈출지 예측해 보세요.

```python
count = 5
while True:
    count -= 1
    if count == 2:
        break
print(count)
```

* **예상 결과:** 2 (이유: count 값이 반복될 때마다 1씩 줄어들다가, if count == 2:라는 조건을 만족하는 순간 break를 만나 반복문이 즉시 종료되기 때문입니다.)

````python
count = 5
while True:
    count -= 1
    if count == 2:
        break
print(count)
````

**문제 5. 반복 건너뛰기 (continue)**
이 코드는 1부터 5까지 반복하면서 어떤 숫자들만 출력할까요?

---

````python
for n in range(1, 6):
    if n % 2 == 0:
        continue
    print(n, end=" ")
````

**미션 2. for문을 이용한 리스트 데이터 일괄 변환**

* **상황:** `prices = [1000, 2000, 3000]` 이라는 원가 리스트가 있습니다. 이 리스트의 모든 요소에 부가세 10%를 더한(즉, 1.1을 곱한) 새로운 리스트를 `for`문을 사용해 만들고 싶습니다.
* **나의 프롬프트:** > "파이썬으로 prices = [1000, 2000, 3000] 리스트를 만들고, 빈 리스트인 new_prices를 생성해 줘. 그다음 for문을 사용해서 prices의 각 요소에 1.1을 곱한 값을 new_prices에 하나씩 추가(append)하고, 마지막에 전체 결과를 출력해 줘.

````python
prices = [1000, 2000, 3000]
new_prices = []

for price in prices:
    new_prices.append(price * 1.1)

print(new_prices)
````

**미션 3. 특정 조건에 맞는 데이터만 필터링 (If + For 결합)**

* **상황:** `numbers = [15, 3, 28, 7, 10, 32]` 리스트가 있습니다. 이 중에서 '10보다 크면서 짝수인 숫자'만 골라서 빈 리스트에 `append`하는 코드를 짜고 싶습니다.
* **나의 프롬프트:** > "파이썬으로 numbers = [15, 3, 28, 7, 10, 32] 리스트를 만들고, 이 중 10보다 크면서 동시에 짝수인 숫자만 골라내고 싶어. for문과 if문을 사용해서 조건에 맞는 숫자들만 result라는 새 리스트에 담아서 출력해 주는 코드를 작성해 줘."

````python
numbers = [15, 3, 28, 7, 10, 32]
result = []

for num in numbers:
    if num > 10 and num % 2 == 0:  # 10보다 크고 짝수인 경우
        result.append(num)

print(result)
````

**미션 4. 리스트 컴프리헨션(List Comprehension)으로 압축하기**

* **상황:** 방금 미션 3에서 여러 줄로 짰던 `for`문과 `if`문 결합 코드를, AI에게 요구하여 파이썬 특유의 간결한 문법인 '리스트 컴프리헨션(한 줄 코딩)'으로 변환하고 싶습니다.
* **나의 프롬프트:** > "방금 작성한 numbers 리스트에서 10보다 크고 짝수인 숫자만 골라내는 로직을, for문과 if문을 여러 줄 쓰지 말고 파이썬의 '리스트 컴프리헨션' 문법을 사용해서 딱 한 줄로 깔끔하게 바꿔 줘."

````python
numbers = [15, 3, 28, 7, 10, 32]
result = [num for num in numbers if num > 10 and num % 2 == 0]

print(result)
````

**미션 5. while문을 활용한 입력 검증 로직**

* **상황:** 사용자에게 `input()`으로 비밀번호를 입력받습니다. 비밀번호가 "python123"이 아니면 계속해서 다시 입력하라고 요구하고, 맞으면 "로그인 성공"을 띄운 뒤 반복을 멈추게 하고 싶습니다.
* **나의 프롬프트:** > "파이썬으로 비밀번호를 확인하는 프로그램을 만들어 줘. while True:를 사용해서 무한 루프를 만들고, 사용자에게 input()으로 비밀번호를 입력받아 봐. 입력한 값이 'python123'이면 '로그인 성공'을 출력하고 break로 멈춰 주고, 아니면 '다시 시도하세요'를 출력하며 계속 입력받게 해 줘."

---

````python
while True:
    password = input("비밀번호를 입력하세요: ")

    if password == "python123":
        print("로그인 성공")
        break
    else:
        print("다시 시도하세요")
````

## 🚀 PART 4. [심화 문제] 스무고개 숫자 맞추기 (Up & Down 게임)

*지금까지 배운 변수, 조건문(if), 반복문(while), 입출력(input, print), 그리고 난수 생성(random)을 모두 융합하여 미니 프로그램을 완성하는 종합 알고리즘 실습입니다.*

**[알고리즘 설계도 (Flowchart)]**

1. 컴퓨터가 `random.randint(1, 50)`을 사용해 1부터 50 사이의 정답 숫자를 하나 생성하여 변수에 저장합니다.
2. 사용자가 정답을 맞힐 때까지 **계속해서(while True)** 숫자를 입력(`input`)받습니다. (입력받은 값은 정수형 `int`로 변환해야 합니다.)
3. 만약 사용자가 입력한 숫자가 정답보다 **크다면**, "Down!"을 출력합니다.
4. 만약 사용자가 입력한 숫자가 정답보다 **작다면**, "Up!"을 출력합니다.
5. 만약 사용자가 입력한 숫자와 정답이 **같다면**, "정답입니다!"를 출력하고 **반복문을 즉시 종료(break)** 합니다.

**[실습 미션]**
게임을 만들기 위하여 AI에게 적절한 프롬프트를 전달하여 AI가 짜준 코드를 Colab에 붙여넣고 직접 게임을 플레이하며 로직이 정상적으로 작동하는지 검증합니다.

* **💻 AI가 작성해 준 코드를 검토하고 실행하기.**

````python
import random

answer = random.randint(1, 50)

while True:
    user_input = int(input("1부터 50 사이의 숫자를 맞혀 보세요: "))

    if user_input > answer:
        print("Down!")
    elif user_input < answer:
        print("Up!")
    else:
        print("정답입니다!")
        break
````

### 설명

* `random.randint(1, 50)` → 1~50 사이의 정답 숫자 생성
* `while True` → 정답을 맞힐 때까지 반복
* `int(input(...))` → 입력값을 정수로 변환
* 입력값이 크면 `"Down!"`, 작으면 `"Up!"` 출력
* 같으면 `"정답입니다!"` 출력 후 `break`로 종료
