### 👋 오늘의 학습 목표

* **결측치(Missing Value):** 데이터에 뻥 뚫린 구멍(비어있는 값)을 찾아내고, 지우거나 다른 값으로 메꿀 수 있다.
* **이상치(Outlier):** 나이가 999살이거나 가격이 -500원인 '말도 안 되는 튀는 값'을 발견하고 제거할 수 있다.
* **데이터 타입(Type) 변경:** 엑셀에서 넘어온 "1,000원"이라는 글자(문자)를 계산이 가능한 진짜 숫자(정수)로 바꿀 수 있다.

## 🎯 PART 1. 핵심 전공 개념 확인

*지저분한 데이터를 분석하기 좋게 청소하는 판다스(Pandas) 필수 용어들입니다.*

1. 수집된 데이터의 헛점과 오류를 찾아내고 수정하여 분석하기 좋은 상태로 만드는 모든 청소 과정을 데이터 **( 전처 )**(Preprocessing)이라고 부릅니다.
2. 데이터프레임 안에서 값이 비어있는 부분, 즉 '결측치'를 판다스에서는 알파벳 3글자로 **( NaN )**(Not a Number)이라고 표시합니다.
3. 데이터프레임에서 비어있는 값(결측치)이 있는지 없는지 찾아내는(True/False로 알려주는) 메서드는 `( isnull )` 입니다.
4. 보통 3번의 함수 뒤에 `.sum()`을 붙여서 `df.isnull().sum()` 형태로 사용하면, 각 열(컬럼)마다 비어있는 데이터가 **총 몇 개**인지 한눈에 세어볼 수 있습니다.
5. 데이터가 너무 많이 비어있어서 쓸모없는 행(Row)을 **통째로 삭제**해 버릴 때 사용하는 메서드는 `( dropna )` 입니다.
6. 비어있는 값을 삭제하는 대신, 0이나 '평균값' 등 **내가 원하는 특정 값으로 채워 넣을 때** 사용하는 메서드는 `( fillna )` 입니다.
7. 나이가 999살이거나 시험 점수가 500점인 것처럼, 정상적인 범위를 크게 벗어난 '말도 안 되는 튀는 값'을 **( 이상치 )**(Outlier)이라고 부릅니다.
8. 가격 열에 "1,000원"이라고 적혀있으면 컴퓨터는 이를 숫자가 아니라 **( object )**(문자열)로 인식하여 더하기, 빼기 같은 계산을 할 수 없습니다.
9. 8번과 같은 글자에서 쉼표(,)나 '원' 같은 특정 글자를 **찾아서 다른 글자로 교체(또는 없애기)** 할 때 사용하는 문자열 메서드는 `str.( replace )` 입니다.
10. 글자 안에 있는 불필요한 기호를 모두 없앤 후, 문자로 인식되던 데이터를 진짜 **정수형 숫자**로 변형(타입 변경)할 때 사용하는 메서드는 `( astype )` 입니다. (예: `astype(int)`)

## 3. 타이타닉 데이터셋

**과제명: 타이타닉호의 비밀 (나이 채우기)**

1. **데이터:** Seaborn의 `titanic` 데이터를 불러옵니다.
2. **문제:** 이 데이터의 `age`(나이) 컬럼에는 약 177개의 결측치가 있습니다.
3. **미션:**
* 그냥 삭제하지 말고, **생존자들의 평균 나이**를 구해서 그 값으로 비어있는 나이를 채워주세요.
* 채우기 전 결측치 개수와, 채운 후 결측치 개수(0개여야 함)를 출력한 화면을 제출하세요.
* Z-score가 ±3을 벗어난 값 찾아아서 제거하세요
* IQR 기준 이상치(Q1 - 1.5IQR 이하 또는 Q3 + 1.5IQR)를 찾아 보세요.
* 나이에 있는 이상치를 어떻게 처리하면 좋을지 알아보세요

**미션 1. 생존자 평균 나이로 결측치 채우기**
* **나의 프롬프트:**
> "seaborn의 titanic 데이터에서 age 컬럼의 결측치 개수를 확인하고, 생존자(survived=1)의 평균 나이로 결측치를 채우는 코드를 알려줘."

````python
import seaborn as sns
import pandas as pd

df = sns.load_dataset('titanic')

# 채우기 전 결측치 개수
print(df['age'].isna().sum())

# 생존자 평균 나이
mean_age = df[df['survived'] == 1]['age'].mean()

# 결측치 채우기
df['age'] = df['age'].fillna(mean_age)
````

**미션 2. Z-score 기반 이상치 제거 (±3)**
* **나의 프롬프트:**
> "타이타닉 데이터에서 age 컬럼의 Z-score를 계산하고 ±3을 벗어나는 이상치를 찾는 코드를 알려줘."

````python
import numpy as np

z = (df['age'] - df['age'].mean()) / df['age'].std()

outliers_z = df[(z > 3) | (z < -3)]

print(len(outliers_z))
````

**미션 3. IQR 기반 이상치 탐지**
* **나의 프롬프트:**
> "타이타닉 데이터의 age 컬럼에서 IQR 방식으로 이상치(Q1 - 1.5IQR 이하 또는 Q3 + 1.5IQR 이상)를 찾는 코드를 알려줘."

````python
Q1 = df['age'].quantile(0.25)
Q3 = df['age'].quantile(0.75)
IQR = Q3 - Q1

lower = Q1 - 1.5 * IQR
upper = Q3 + 1.5 * IQR

outliers_iqr = df[(df['age'] < lower) | (df['age'] > upper)]

print(len(outliers_iqr))
````

**미션 4. 나이 이상치 처리 방법 정리**
* **나의 프롬프트:**
> "`titanic` 데이터의 `age` 이상치를 어떻게 처리하는 것이 좋은지 방법을 설명해 줘."

**방법 1. 제거**

```python
df_clean = df[(df['age'] >= lower) & (df['age'] <= upper)]
```

---

**방법 2. 경계값으로 보정 (Winsorizing)**

```python
df['age'] = np.where(df['age'] > upper, upper, df['age'])
df['age'] = np.where(df['age'] < lower, lower, df['age'])
```
