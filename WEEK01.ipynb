### 🚩 미션 2: 나만의 계산기 만들기

수학 계산은 컴퓨터가 제일 잘합니다. AI에게 복잡한 계산을 시켜봅시다.

> **프롬프트 입력:**
> "파이썬으로 현재 시간을 기준으로 내가 태어난 지 며칠이나 지났는지 계산해주는 코드를 작성해줘. 내 생일은 [YYYY년 MM월 DD일]이야."

**[결과 확인]**

* 나는 태어난 지 약 7954일이 지났습니다.

`````python
from datetime import date

# 생일 설정
birthday = date(2004, 6, 23)

# 현재 날짜
today = date.today()

# 태어난 후 지난 일수 계산
days_lived = (today - birthday).days

# 결과 출력
print(f"나는 태어난 지 약 {days_lived}일이 지났습니다.")
`````
