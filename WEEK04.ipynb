**미션 2. 복잡한 다중 조건 필터링 (AND, OR 연산)**

* **상황:** `students` 데이터프레임에서 전공(`major`)이 '컴퓨터공학'이면서(AND), 학점(`gpa`)이 4.0 이상인 학생들만 추출하여 새로운 데이터프레임 `honor_students`를 만들고 싶습니다.
* **나의 프롬프트:** > "파이썬 판다스를 사용해서 `students` 데이터프레임에서 전공(`major`)이 '컴퓨터공학'이면서 동시에(AND) 학점(`gpa`)이 4.0 이상인 학생들만 필터링하고, 그 결과를 `honor_students`라는 새로운 데이터프레임으로 저장하는 코드를 작성해 줘."

````python
# @title
import pandas as pd
import random
import numpy as np

# 시드 고정
random.seed(42)
np.random.seed(42)

# 학생 수
n = 100

# 학번: 8자리, 앞 4자리는 연도(예: 2026)
ids = [f"2026{str(i).zfill(4)}" for i in range(1, n+1)]

# 전공: "AI", "CS", "EE" 랜덤 배치
majors = [random.choice(["AI", "CS", "EE"]) for _ in range(n)]

# 성별: "M", "F" 랜덤 배치
sexes = [random.choice(["M", "F"]) for _ in range(n)]

# 학점: 평균 3.5, 표준편차 0.5, 범위 1.0 ~ 4.5
gpas = np.random.normal(loc=3.5, scale=0.5, size=n)
gpas = np.clip(gpas, 1.0, 4.5)

# 소수점 2자리로 반올림
gpas = [round(g, 2) for g in gpas]

# 데이터 프레임 생성
students = pd.DataFrame({
    "id": ids,
    "major": majors,
    "sex": sexes,
    "gpa": gpas
})

print(students.head())  # 앞부분 확인
````

````python
# 전공 "CS"를 "컴퓨터공학"으로 바꾸기 (실제 조건 맞추기)
students['major'] = students['major'].replace('CS', '컴퓨터공학')

# 필터링: 전공이 '컴퓨터공학'이면서 학점이 4.0 이상
honor_students = students[(students['major'] == '컴퓨터공학') & (students['gpa'] >= 4.0)]

# 결과 확인
honor_students.head()
````

**미션 3. 특정 열(Column) 기준 정렬하기**

* **상황:** `sales` 데이터프레임이 있습니다. 이 표를 매출액(`revenue`) 열을 기준으로 가장 높은 숫자부터 낮아지는 순서(내림차순)로 정렬하여 상위 10개만 보고 싶습니다.
* **나의 프롬프트:** > "파이썬 판다스를 사용해서 `sales` 데이터프레임을 매출액(`revenue`) 열을 기준으로 내림차순으로 정렬한 뒤, 상위 10개 행만 추출하는 코드를 작성해 줘."

````python
# @title
import pandas as pd
import numpy as np

# 시드 고정 (재현 가능)
np.random.seed(42)

# 아이템 개수
n = 10

# 아이템 이름
items = [f"Item{i+1}" for i in range(n)]

# 단가: 1000 ~ 5000 사이 랜덤
prices = np.random.randint(1000, 5001, size=n)

# 수량: 1 ~ 20 사이 랜덤
quantities = np.random.randint(1, 21, size=n)

# 매출액 = 단가 * 수량
revenues = prices * quantities

# 데이터프레임 생성
sales = pd.DataFrame({
    "item": items,
    "price": prices,
    "quantity": quantities,
    "revenue": revenues
})

print(sales.head())
````

````python
# 매출액(revenue) 기준 내림차순 정렬 후 상위 10개
top10_sales = sales.sort_values(by='revenue', ascending=False).head(10)

# 결과 확인
top10_sales
````

**미션 4. 결측치(Missing Values) 처리하기**

* **상황:** `df` 데이터프레임에 비어있는 값(NaN)이 너무 많습니다. 어느 열에 결측치가 몇 개씩 있는지 확인하는 코드와, 결측치가 하나라도 포함된 행(Row)을 모두 삭제해 버리는 코드를 알고 싶습니다.
* **나의 프롬프트:** > "파이썬 판다스를 사용해서 `df` 데이터프레임의 각 열별 결측치(NaN) 개수를 확인하는 코드와, 결측치가 하나라도 포함된 모든 행을 삭제하는 코드를 각각 작성해 줘."

````python
# @title
import pandas as pd
import numpy as np

# 시드 고정
np.random.seed(42)

# 아이템 개수
n = 10

# 아이템 이름
items = [f"Item{i+1}" for i in range(n)]

# 단가: 1000 ~ 5000 사이 랜덤
prices = np.random.randint(1000, 5001, size=n)

# 수량: 1 ~ 20 사이 랜덤
quantities = np.random.randint(1, 21, size=n)

# 매출액 = 단가 * 수량
revenues = prices * quantities

# 데이터프레임 생성
df = pd.DataFrame({
    "item": items,
    "price": prices,
    "quantity": quantities,
    "revenue": revenues
})

# 결측치 4개 삽입 (예: price, quantity, revenue에 랜덤하게)
df.loc[0, "price"] = np.nan
df.loc[3, "quantity"] = np.nan
df.loc[5, "revenue"] = np.nan
df.loc[7, "price"] = np.nan

print(df.info())
````

````python
import pandas as pd
import numpy as np

# 1️⃣ 각 열별 결측치 개수 확인
missing_counts = df.isna().sum()
print("열별 결측치 개수:\n", missing_counts)

# 2️⃣ 결측치가 하나라도 있는 행 삭제
result_df = df.dropna()

# 결과 확인
print("\n결측치 제거 후 데이터프레임:\n", result_df)
````

3. 조건 검색 (VIP 구출 작전):
'titanic' 데이터프레임에서 객실 등급(pclass)이 1등급(1)이고,
나이(age)가 30세 미만이며, 생존(survived)한 승객만 추려서 'vip_survivors'라는 새로운 데이터프레임에 저장하는 코드를 작성해 줘."

4. 결과 확인:
"'vip_survivors' 데이터프레임의 총 승객 수를 화면에 출력하는 코드를 작성해 줘."

````python
import pandas as pd
import seaborn as sns

# 시본 내장 타이타닉 데이터셋 불러오기
titanic = sns.load_dataset('titanic')

# 데이터 확인 (선택)
titanic.head()
````

````python
# 데이터 크기 확인 (행, 열)
print("데이터 크기 (행, 열):", titanic.shape)

# 열별 결측치 개수 확인
print("\n열별 결측치 개수:\n", titanic.isna().sum())

# 각 열 데이터 타입 확인
print("\n데이터 타입:\n", titanic.dtypes)
````

````python
# 조건 필터링: 1등급 객실, 나이 30세 미만, 생존한 승객
vip_survivors = titanic[
    (titanic['pclass'] == 1) &
    (titanic['age'] < 30) &
    (titanic['survived'] == 1)
]

# 결과 확인 (선택)
vip_survivors.head()
````

````python
# vip_survivors의 총 행 개수 → 총 승객 수
total_vip = len(vip_survivors)

print("살아남은 1등급 젊은 승객 수:", total_vip)
````
