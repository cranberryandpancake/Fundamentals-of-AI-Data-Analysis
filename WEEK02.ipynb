**미션 2. 리스트 생성과 길이 확인**

* **상황:** 내가 이번 학기에 수강하는 전공과목 이름 4개가 들어간 리스트를 만들고, 데이터가 몇 개인지 확인하고 싶습니다.
* **나의 프롬프트:** > "파이썬으로 "자료 구조", "인공 지능", "인공 지능 데이터 분석 기초", "머신 러닝" 4개의 과목 이름이 들어간 리스트를 생성하는 코드를 작성해 줘. 그리고 len() 함수를 사용해서 리스트에 들어있는 데이터의 개수를 확인하고 출력하는 코드도 함께 작성해 줘."

````python
# 과목 이름이 들어간 리스트 생성
subjects = ["자료 구조", "인공 지능", "인공 지능 데이터 분석 기초", "머신 러닝"]

# 리스트에 들어있는 데이터의 개수 확인
count = len(subjects)

# 결과 출력
print("과목 리스트:", subjects)
print("리스트에 들어있는 데이터의 개수:", count)
````

**미션 3. 특정 위치에 데이터 삽입하기 (`insert`)**

* **상황:** 미션 2에서 만든 리스트의 **맨 앞(0번 인덱스)** 에 '파이썬 프로그래밍'이라는 과목을 새롭게 끼워 넣고 싶습니다.
* **나의 프롬프트:** > "파이썬으로 `subjects` 리스트에 insert() 메서드를 사용해서 **0번 인덱스(맨 앞)**에 "파이썬 프로그래밍"이라는 과목을 추가하는 코드를 작성해 줘. 그리고 과목이 추가된 후의 리스트를 출력하는 코드도 함께 작성해 줘."

````python
# 0번 인덱스(맨 앞)에 과목 추가
subjects.insert(0, "파이썬 프로그래밍")

# 결과 출력
print("과목이 추가된 리스트:", subjects)
````

**미션 4. 리스트 슬라이싱 활용하기**

* **상황:** 전공과목 리스트에 과목이 5개로 늘어났습니다. 이 중에서 **앞에서부터 딱 3과목만** 잘라내서 새로운 리스트로 저장하고 싶습니다.
* **나의 프롬프트:** > "파이썬으로 `subjects` 리스트에 리스트 슬라이싱을 사용해서 앞에서부터 3개의 과목만 잘라 새로운 리스트로 저장하는 코드를 작성해 줘. 그리고 새로 생성된 리스트를 출력하는 코드도 함께 작성해 줘."

````python
# 앞에서부터 3개의 과목만 슬라이싱하여 새로운 리스트 생성
new_subjects = subjects[:3]

# 결과 출력
print("새로 생성된 리스트:", new_subjects)
````

**미션 5. 리스트 내림차순 정렬 (`sort` 활용)**

* **상황:** `scores = [85, 92, 78, 100, 88]` 이라는 점수 리스트가 있습니다. 이 점수들을 **가장 높은 점수부터 낮아지는 순서(내림차순)** 로 정렬하고 싶습니다.
* **나의 프롬프트:** > "파이썬으로 `scores = [85, 92, 78, 100, 88]` 리스트가 있을 때, sort() 메서드를 사용해서 점수를 높은 순서(내림차순)로 정렬하는 코드를 작성해 줘. 그리고 정렬된 리스트를 출력하는 코드도 함께 작성해 줘."

````python
# 점수 리스트
scores = [85, 92, 78, 100, 88]

# 내림차순으로 정렬
scores.sort(reverse=True)

# 결과 출력
print("내림차순으로 정렬된 점수 리스트:", scores)
````

**미션 6. [데이터 전처리 예습] 문자열을 리스트로 쪼개기 (`split`)**

* **상황:** `"사과,바나나,포도,수박"` 처럼 쉼표(,)로 연결된 하나의 긴 문자열이 있습니다. 이것을 쉼표를 기준으로 쪼개서 4개의 과일이 들어간 **리스트**로 변환하고 싶습니다.
* **나의 프롬프트:** > "파이썬으로 "사과,바나나,포도,수박"이라는 문자열이 있을 때, split() 메서드를 사용해서 쉼표(,)를 기준으로 문자열을 나누어 4개의 과일이 들어있는 리스트로 변환하는 코드를 작성해 줘. 그리고 변환된 리스트를 출력하는 코드도 함께 작성해 줘."
---

````python
# 문자열 생성
fruits_str = "사과,바나나,포도,수박"

# 쉼표를 기준으로 문자열을 나누어 리스트로 변환
fruits_list = fruits_str.split(",")

# 결과 출력
print("리스트로 변환된 과일:", fruits_list)
````
